```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
X,y = make_blobs(centers=4, n_samples=200, random_state=0, cluster_std=0.7)
print(X[:10],y[:10])
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.cluster import KMeans
model = KMeans(4)
model.fit(X)
print(model.cluster_centers_)
```

```python
plt.scatter(X[:,0],X[:,1], c=model.labels_);
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:,1], s=100, color="red"); # Show the centres
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y, model.labels_)
print("Accuracy score is", acc)
```

```python
def find_permutation(n_clusters, real_labels, labels):
    permutation = []
    for i in range(n_clusters):
        idx = labels == i
        if np.any(idx):  # Ensure there are data points in the cluster
            mode_result = scipy.stats.mode(real_labels[idx])  # Get the mode result object
            new_label = mode_result.mode  # Access the mode value safely
            permutation.append(new_label)
    return permutation
permutation = find_permutation(4, y, model.labels_)
print(permutation)
```

```python
new_labels = [ permutation[label] for label in model.labels_]   # permute the labels
print("Accuracy score is", accuracy_score(y, new_labels))
```
